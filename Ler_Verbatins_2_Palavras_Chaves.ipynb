```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import spacy

#!python -m spacy download pt_core_news_sm
```

```python
# Carregar modelo de portugu√™s do spaCy
nlp = spacy.load("pt_core_news_sm")
```

```python
# Carregar os dados
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Arquivos_Ler/Verbatins.xlsx")  # ou read_csv, se for CSV
```

```python
df.head(10)
```

```python
# Instanciar o analisador de sentimentos
analisador = SentimentIntensityAnalyzer()
```

```python
# Fun√ß√£o para an√°lise de sentimento
def analisar_vader(texto):
    resultado = analisador.polarity_scores(str(texto))
    pontuacao = resultado["compound"]
    if pontuacao <= -0.2:
        categoria = "Negativo"
    elif pontuacao >= 0.2:
        categoria = "Positivo"
    else:
        categoria = "Neutro"
    return pd.Series([pontuacao, categoria])
```

```python
# Fun√ß√£o para extrair palavras-chave
def extrair_keywords(texto):
    doc = nlp(str(texto))
    palavras = [token.text.lower() for token in doc if token.pos_ in ["NOUN", "PROPN", "ADJ"] and not token.is_stop]
    return ", ".join(palavras)
```

```python
# Aplicar as fun√ß√µes
df[["Pontuacao", "Categoria"]] = df["Comentarios"].apply(analisar_vader)
df["PalavrasChave"] = df["Comentarios"].apply(extrair_keywords)
```
