```python
#!pip install pandasql
import pandas as pd
from pandasql import sqldf
```

```python
# Ler o arquivo especifico em xls
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Arquivos_Ler/BasesManualParaAcionamentoAutomatico.xlsx')
```

```python
#Remover espaço em colunas
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Consulta SQL diretamente df
Consulta = """
SELECT
      Id_unico
    , matricula
    , nomeagente
    , supervisor
    , loginrede
    , loginavaya
    , loginaspect
    , acessozap
    , acessocacs
    , datacontato
    , tabulacao
    , tipodenegocio
    , eps
    , horainicial
    , horafinal
    , Duracao
    , cpf_cnpj
    , numerodiscado
    , TRIM(substr(numerodiscado, 3)) AS SoNumero
    , substr(numerodiscado, 1, 2) AS DD
    , nome_audio
FROM df
WHERE 1=1
AND supervisor = 'PAULO HENRIQUE MILAN'
"""
resultado = sqldf(Consulta)
```

```python
resultado.head(5)
```

```python
#Converte campos em formato normal data e hora
resultado['datacontato'] = pd.to_datetime(resultado['datacontato'], errors='coerce').dt.date
resultado['horainicial'] = pd.to_datetime(resultado['horainicial'], errors='coerce').dt.strftime('%H:%M:%S')
resultado['horafinal'] = pd.to_datetime(resultado['horafinal'], errors='coerce').dt.strftime('%H:%M:%S')
```

```python
resultado.head(5)
```

```python
resultado.to_csv('/content/BasesManualParaAcionamentoAutomatico.csv', index=False)
```
