```python
#!pip install pandasql
import pandas as pd
from pandasql import sqldf
import time
```

```python
# Ler o arquivo especifico em xls
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Arquivos_Ler/BasesManualParaAcionamentoAutomatico.xlsx')
```

```python
df_feriado = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Arquivos_Ler/Feriados.xlsx')
```

```python
df_feriado.head()
```

```python
df_feriado.dtypes
```

```python
#converte coluna DATA_FERIADO PARA object
df_feriado['DATA_FERIADO'] = df_feriado['DATA_FERIADO'].astype(str)
```

```python
#Pegar a data de hoje
hoje = time.strftime("%Y-%m-%d")
```

```python
#Remover espaço em colunas
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Consulta SQL diretamente df
Consulta = """
SELECT
      Id_unico
    , matricula
    , nomeagente
    , supervisor
    , loginrede
    , loginavaya
    , loginaspect
    , acessozap
    , acessocacs
    , datacontato
    , tabulacao
    , tipodenegocio
    , eps
    , horainicial
    , horafinal
    , Duracao
    , cpf_cnpj
    , numerodiscado
    , TRIM(substr(numerodiscado, 3)) AS SoNumero
    , substr(numerodiscado, 1, 2) AS DD
    , nome_audio
FROM df
WHERE 1=1
AND supervisor = 'PAULO HENRIQUE MILAN'
"""
resultado = sqldf(Consulta)
```

```python
resultado.head(5)
```

```python
#Consulta SQL diretamente df
Consulta1 = """
SELECT
    UPPER(tabulacao) AS "Tabulação",
    COUNT(1) AS "Quantidade",
    ROUND(100.0 * COUNT(1) / (SELECT COUNT(1) FROM df WHERE tabulacao IS NOT NULL), 2) || '%' AS "Representatividade"
FROM df
WHERE tabulacao IS NOT NULL
GROUP BY UPPER(tabulacao)
ORDER BY "Quantidade" DESC;
"""
resultado_sup = sqldf(Consulta1)
```

```python
resultado_sup
```

```python
#Contar linhas e colunas
resultado.shape
```

```python
#Converte campos em formato normal data e hora
resultado['datacontato'] = pd.to_datetime(resultado['datacontato'], errors='coerce').dt.date
resultado['horainicial'] = pd.to_datetime(resultado['horainicial'], errors='coerce').dt.strftime('%H:%M:%S')
resultado['horafinal'] = pd.to_datetime(resultado['horafinal'], errors='coerce').dt.strftime('%H:%M:%S')
```

```python
# Substitui o mês 08 pelo mês 02
resultado['datacontato'] = resultado['datacontato'].astype(str).str.replace(
    r'^(\d{4})-08-(\d{2})$', r'\1-02-\2', regex=True
)
```

```python
resultado.head(5)
```

```python
resultado.dtypes
```

```python
#cruzar resultado com df_feriado
resultadofinal = pd.merge(
    resultado,
    df_feriado,
    left_on='datacontato',
    right_on='DATA_FERIADO',
    how='left'
)
```

```python
resultadofinal.head()
```

```python
resultadofinal.to_csv(f'/content/BasesManualParaAcionamentoAutomatico_{hoje}.csv', index=False)
resultado_sup.to_csv(f'/content/resultado_sup_{hoje}.csv', index=False)
```
