```python
# LEndo um arquivo CSV
df_10 = pd.read_csv('C:\Aplicativos\Meus_Arquivos_Willian\Ambiente_Desenvolvimento\Arquivos_in\Clientes.csv', sep=',')
df_10['Dt_ingest'] = Data_Hoje
df_10['Dt_ingest'] = pd.to_datetime(df_10['Dt_ingest'], format='%d-%m-%Y')
# extrair data da coluna created_at
df_10['created_at_data'] = pd.to_datetime(df_10['created_at']).dt.normalize()


display(df_10)
```

```python
# selecionar colunas específicas
dfcolunas = df_10[['created_at','id', 'email', 'state','Dt_ingest', 'created_at_data']]
# Filtrar os dados na coluna created_at_data
df_intervalo = dfcolunas[
    (dfcolunas['created_at_data'] >= pd.to_datetime('2018-01-01')) &
    (dfcolunas['created_at_data'] < pd.to_datetime(Data_Hoje))
]

df_intervalo.head(500)
```

```python
# Outra meneira de filtrar os dados
Colunas =['created_at','id', 'email', 'state','Dt_ingest', 'created_at_data']
Colunas_Filtradas= df_10[Colunas]
Colunas_Filtro = Colunas_Filtradas[(Colunas_Filtradas['created_at_data'] >= pd.to_datetime('2018-01-01')) & (Colunas_Filtradas['created_at_data'] < pd.to_datetime(Data_Hoje))]

Colunas_Filtro.head(10)
```
