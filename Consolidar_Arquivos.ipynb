Consolidando arquivos de uma pasta

```python
#Diretório onde estão os arquivos
diretorio = r"C:\Aplicativos\Meus_Arquivos_Willian\Ambiente_Desenvolvimento\Arquivos_in\Vendas"

#Lista os arquivos do diretório
arquivos = []

#Iterar sobre os arquivos no diretório
for arquivo in os.listdir(diretorio):                        #lista os arquivos no diretório
    if arquivo.endswith(".xlsx"):                            #Verifica se o arquivo é um Excel
        caminho_completo = os.path.join(diretorio, arquivo)  #Cria o caminho completo do arquivo
        df = pd.read_excel(caminho_completo)                 #Lê o arquivo Excel
        arquivos.append(df)                                  #Adiciona o DataFrame à lista 1 a 1
#Consolidar os DataFrames em um único DataFrame
consolidado = pd.concat(arquivos, ignore_index=True)
consolidado.head(20)
```

```python
#Colacar a data de hoje no arquivo consolidado
consolidado['Dt_ingest'] = Data_Hoje
consolidado.head(20)
```

```python
#Converte as data para formato de data
consolidado['Dt_ingest'] = pd.to_datetime(consolidado['Dt_ingest'], format='%d-%m-%Y')
consolidado['Data'] = pd.to_datetime(consolidado['Data'], format='%d-%m-%Y')
```

```python
#Subtrrair Dt_ingest da Data ver os dias de diferença
consolidado['Diferenca_Dias'] = (consolidado['Dt_ingest'] - consolidado['Data']).dt.days
```

```python
consolidado.head(20)
```

```python
# ver a diferença em meses
consolidado['Diferenca_Meses'] = (consolidado['Dt_ingest'] - consolidado['Data']).dt.days // 30
```

```python
consolidado.head(20)
```

```python
#ver a diferença em anos
consolidado['Diferenca_Ano'] = (consolidado['Dt_ingest'] - consolidado['Data']).dt.days // 365
```

```python
consolidado.head(20)
```

```python
# consolidar dias + meses + anos concatenados
consolidado['Diferenca_Total'] = (
    consolidado['Diferenca_Dias'].astype(str) + ' dias, ' +
    consolidado['Diferenca_Meses'].astype(str) + ' meses, ' +
    consolidado['Diferenca_Ano'].astype(str) + ' anos'
)
consolidado.head(10)
```

Salvando consolidado em uma pasta especifica Formato Excel

```python
# Caminho da pasta onde estão os arquivos
nome_arquivo = "Dados consolidado Vendas "
caminho_pasta = r"C:\Aplicativos\Meus_Arquivos_Willian\Ambiente_Desenvolvimento\Arquivos_out\Vendas\Consolidado"
# Verifica se a pasta existe, se não existir, cria a pasta
if not os.path.exists(caminho_pasta):
    os.makedirs(caminho_pasta)
# Caminho completo do arquivo de saída
caminho_arquivo_saida = os.path.join(caminho_pasta, f"{nome_arquivo}{Data_Hoje}.xlsx")
# Salva o DataFrame consolidado em um arquivo Excel
consolidado.to_excel(caminho_arquivo_saida, index=False)
```

```python
#Abrir o diretório onde foi salvo o arquivo
os.startfile(caminho_pasta) # Abrir o diretório onde foi salvo o arquivo
print(f"Arquivo consolidado salvo em: {caminho_arquivo_saida}")
print("Processo de consolidação concluído com sucesso!")
```
