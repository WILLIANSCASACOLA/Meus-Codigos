```python
import os
from datetime import datetime, timedelta

# Mapeamento manual dos dias da semana em portuguÃªs
dias_semana_pt = {
    0: "segunda-feira",
    1: "terÃ§a-feira",
    2: "quarta-feira",
    3: "quinta-feira",
    4: "sexta-feira",
    5: "sÃ¡bado",
    6: "domingo"
}

# Feriados nacionais fixos com nomes
feriados_fixos = {
    "01/01": "ConfraternizaÃ§Ã£o Universal",
    "07/09": "IndependÃªncia do Brasil",
    "12/10": "Nossa Senhora Aparecida",
    "02/11": "Finados",
    "25/12": "Natal"
}

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def saudacao_inicial():
    agora = datetime.now()
    hora = agora.hour
    if hora < 12:
        saudacao = "Bom dia"
    elif hora < 18:
        saudacao = "Boa tarde"
    else:
        saudacao = "Boa noite"
    print(f"{saudacao}, seja bem-vindo Ã  Calculadora de Prazo em Dias Ãšteis!")
    print(f"ðŸ“… Hoje Ã© {agora.strftime('%d/%m/%Y')} ({dias_semana_pt[agora.weekday()]})")
    print(f"ðŸ•’ Hora atual: {agora.strftime('%H:%M')}\n")

def mostrar_feriados():
    print("\nðŸ“… Feriados considerados na contagem:")
    for data, nome in feriados_fixos.items():
        print(f" - {data} â†’ {nome}")

def is_dia_util(data):
    if data.weekday() >= 5:
        return False
    if data.strftime("%d/%m") in feriados_fixos:
        return False
    return True

def calcular_prazo(data_inicial, dias_uteis, considerar_dia_solicitacao):
    data = data_inicial
    if not considerar_dia_solicitacao:
        data += timedelta(days=1)

    dias_contados = 0
    dias_total = 0
    sabados = 0
    domingos = 0
    feriados = 0
    ignorados = []

    while dias_contados < dias_uteis:
        tipo = None
        if is_dia_util(data):
            dias_contados += 1
        else:
            if data.weekday() == 5:
                sabados += 1
                tipo = "sÃ¡bado"
            elif data.weekday() == 6:
                domingos += 1
                tipo = "domingo"
            elif data.strftime("%d/%m") in feriados_fixos:
                feriados += 1
                tipo = f"feriado - {feriados_fixos[data.strftime('%d/%m')]}"
            if tipo:
                ignorados.append(f"{data.strftime('%d/%m/%Y')} ({tipo})")
        data += timedelta(days=1)
        dias_total += 1

    data_final = data - timedelta(days=1)
    return data_final, dias_total, sabados, domingos, feriados, ignorados

# ðŸ§‘â€ðŸ’» Loop principal
while True:
    limpar_tela()
    saudacao_inicial()
    mostrar_feriados()

    print("\nDeseja usar a data de hoje ou digitar uma data?")
    print("1 - Usar a data de hoje")
    print("2 - Digitar uma data manualmente")
    escolha_data = input("Escolha 1 ou 2: ")

    if escolha_data == "1":
        data_solicitacao = datetime.today()
        nome_dia = dias_semana_pt[data_solicitacao.weekday()]
        print(f"\nðŸ“… Data selecionada automaticamente: {data_solicitacao.strftime('%d/%m/%Y')} ({nome_dia})")
    elif escolha_data == "2":
        entrada_data = input("Digite a data da solicitaÃ§Ã£o (formato DD/MM/AAAA): ")
        try:
            data_solicitacao = datetime.strptime(entrada_data, "%d/%m/%Y")
            nome_dia = dias_semana_pt[data_solicitacao.weekday()]
            print(f"\nðŸ“… Data digitada: {data_solicitacao.strftime('%d/%m/%Y')} ({nome_dia})")
        except ValueError:
            print("âŒ Data invÃ¡lida. Use o formato DD/MM/AAAA.")
            input("\nPressione ENTER para tentar novamente...")
            continue
    else:
        print("âŒ OpÃ§Ã£o invÃ¡lida.")
        input("\nPressione ENTER para tentar novamente...")
        continue

    try:
        dias_uteis = int(input("\nDigite o nÃºmero de dias Ãºteis para o prazo: "))
    except ValueError:
        print("âŒ NÃºmero de dias invÃ¡lido.")
        input("\nPressione ENTER para tentar novamente...")
        continue

    print("\nComo deseja iniciar a contagem?")
    print("1 - Considerar a data da solicitaÃ§Ã£o")
    print("2 - ComeÃ§ar no prÃ³ximo dia Ãºtil")
    opcao = input("Escolha 1 ou 2: ")

    if opcao == "1":
        considerar_dia_solicitacao = True
        tipo_contagem = "Considerando a data da solicitaÃ§Ã£o"
    elif opcao == "2":
        considerar_dia_solicitacao = False
        tipo_contagem = "ComeÃ§ando no prÃ³ximo dia Ãºtil"
    else:
        print("âŒ OpÃ§Ã£o invÃ¡lida.")
        input("\nPressione ENTER para tentar novamente...")
        continue

    prazo_final, dias_corridos, sabados, domingos, feriados, ignorados = calcular_prazo(
        data_solicitacao, dias_uteis, considerar_dia_solicitacao
    )
    nome_dia_final = dias_semana_pt[prazo_final.weekday()]
    print(f"\nâœ… Prazo final para resoluÃ§Ã£o: {prazo_final.strftime('%d/%m/%Y')} ({nome_dia_final})")

    print("\nðŸ“Š EstatÃ­sticas da contagem:")
    print(f" - Data da solicitaÃ§Ã£o: {data_solicitacao.strftime('%d/%m/%Y')} ({dias_semana_pt[data_solicitacao.weekday()]})")
    print(f" - Tipo de contagem: {tipo_contagem}")
    print(f" - Dias Ãºteis solicitados: {dias_uteis}")
    print(f" - Data final: {prazo_final.strftime('%d/%m/%Y')} ({nome_dia_final})")
    print(f" - Dias corridos entre inÃ­cio e fim: {dias_corridos}")
    print(f" - Dias nÃ£o Ãºteis ignorados:")
    print(f"    â€¢ {sabados} sÃ¡bado(s)")
    print(f"    â€¢ {domingos} domingo(s)")
    print(f"    â€¢ {feriados} feriado(s)")

    if ignorados:
        print("\nðŸ“… Dias ignorados na contagem:")
        for dia in ignorados:
            print(f"   - {dia}")

    agora = datetime.now()
    print(f"\nâœ… Consulta executada com Ãªxito em {agora.strftime('%d/%m/%Y')} Ã s {agora.strftime('%H:%M')}")

    voltar = input("\nDeseja fazer outra consulta? (S/N): ").strip().upper()
    if voltar != "S":
        print("\nðŸ‘‹ Encerrando o programa. AtÃ© a prÃ³xima!")
        break
```
