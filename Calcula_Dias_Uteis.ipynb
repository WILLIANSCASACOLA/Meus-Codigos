```python
import os
from datetime import datetime, timedelta

# Feriados nacionais fixos considerados
feriados_fixos = [
    "01/01",  # Confraterniza√ß√£o Universal
    "07/09",  # Independ√™ncia do Brasil
    "12/10",  # Nossa Senhora Aparecida
    "02/11",  # Finados
    "25/12",  # Natal
]

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def mostrar_feriados():
    print("\nüìÖ Feriados considerados na contagem:")
    for f in feriados_fixos:
        print(f" - " + f)

def is_dia_util(data):
    if data.weekday() >= 5:  # s√°bado (5) ou domingo (6)
        return False
    if data.strftime("%d/%m") in feriados_fixos:
        return False
    return True

def calcular_prazo(data_inicial, dias_uteis, considerar_dia_solicitacao):
    data = data_inicial
    if not considerar_dia_solicitacao:
        data += timedelta(days=1)  # come√ßa no pr√≥ximo dia √∫til

    dias_contados = 0
    while dias_contados < dias_uteis:
        if is_dia_util(data):
            dias_contados += 1
        data += timedelta(days=1)

    return data - timedelta(days=1)

# üßë‚Äçüíª Loop principal
while True:
    limpar_tela()
    print("üìå Calculadora de prazo em dias √∫teis")
    mostrar_feriados()

    # Escolha da data de solicita√ß√£o
    print("\nDeseja usar a data de hoje ou digitar uma data?")
    print("1 - Usar a data de hoje")
    print("2 - Digitar uma data manualmente")
    escolha_data = input("Escolha 1 ou 2: ")

    if escolha_data == "1":
        data_solicitacao = datetime.today()
        print(f"\nüìÖ Data selecionada automaticamente: {data_solicitacao.strftime('%d/%m/%Y')} ({data_solicitacao.strftime('%A')})")
    elif escolha_data == "2":
        entrada_data = input("Digite a data da solicita√ß√£o (formato DD/MM/AAAA): ")
        try:
            data_solicitacao = datetime.strptime(entrada_data, "%d/%m/%Y")
            print(f"\nüìÖ Data digitada: {data_solicitacao.strftime('%d/%m/%Y')} ({data_solicitacao.strftime('%A')})")
        except ValueError:
            print("‚ùå Data inv√°lida. Use o formato DD/MM/AAAA.")
            input("\nPressione ENTER para tentar novamente...")
            continue
    else:
        print("‚ùå Op√ß√£o inv√°lida.")
        input("\nPressione ENTER para tentar novamente...")
        continue

    # Entrada do n√∫mero de dias √∫teis
    try:
        dias_uteis = int(input("\nDigite o n√∫mero de dias √∫teis para o prazo: "))
    except ValueError:
        print("‚ùå N√∫mero de dias inv√°lido.")
        input("\nPressione ENTER para tentar novamente...")
        continue

    # Escolha de contagem
    print("\nComo deseja iniciar a contagem?")
    print("1 - Considerar a data da solicita√ß√£o")
    print("2 - Come√ßar no pr√≥ximo dia √∫til")
    opcao = input("Escolha 1 ou 2: ")

    if opcao == "1":
        considerar_dia_solicitacao = True
    elif opcao == "2":
        considerar_dia_solicitacao = False
    else:
        print("‚ùå Op√ß√£o inv√°lida.")
        input("\nPressione ENTER para tentar novamente...")
        continue

    # C√°lculo final
    prazo_final = calcular_prazo(data_solicitacao, dias_uteis, considerar_dia_solicitacao)
    print(f"\n‚úÖ Prazo final para resolu√ß√£o: {prazo_final.strftime('%d/%m/%Y')} ({prazo_final.strftime('%A')})")

    # Pergunta se o usu√°rio quer consultar outro
    voltar = input("\nDeseja fazer outra consulta? (S/N): ").strip().upper()
    if voltar != "S":
        print("\nüëã Encerrando o programa. At√© a pr√≥xima!")
        break
```
