Conexão com API do IBGE

Metodos da API

GET - Solicitar dados
POST - Enviar Dados
PUT - Atualizar um recurso existente, ou atualizar um recurso se ele não exixtir
DELETE - Excluir
PATCH - Aplicar modificações parciais a um recurso

```python
# Configurar opções de exibicção para mostrar as colunas  e linhas commpletas
pd.set_option('display.max_columns', None) # Mostrar todas as colunas
pd.set_option('display.max_rows', None) # Mostrar todas as linhas
pd.set_option('display.max_colwidth', None) # Mostrar o conteúdo completo das colunas nunca truncar os dados


URL="servicodados.ibge.gov.br/api/v1/localidades/distritos"
response = requests.get(f"https://{URL}")
data = response.json()

df3=pd.DataFrame(data)
df3.head()
```

```python
#Usando o recurso de normailizar dados
df3=pd.json_normalize(data)
df3.head()
```
