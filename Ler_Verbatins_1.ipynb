```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
#Ler verbatins
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Arquivos_Ler/Verbatins.xlsx")
```

```python
df.head(10)
```

```python
# Instanciar o analisador
analisador = SentimentIntensityAnalyzer()
```

```python
# Função para analisar o sentimento e categorizar
def analisar_vader(texto):
    resultado = analisador.polarity_scores(str(texto))
    pontuacao = resultado["compound"]  # varia de -1 a 1
    if pontuacao <= -0.2:
        categoria = "Negativo"
    elif pontuacao >= 0.2:
        categoria = "Positivo"
    else:
        categoria = "Neutro"
    return pd.Series([pontuacao, categoria])
```

```python
# Aplicar ao DataFrame
df[["Pontuacao", "Categoria"]] = df["Comentarios"].apply(analisar_vader)
```

```python
# Exportar o resultado
df.to_csv("/content/drive/MyDrive/Colab Notebooks/Arquivos_Ler/comentarios_analisados_vader.csv", index=False)
```
